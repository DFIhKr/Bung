# 3. Upscale video

- The upscaled video will be saved to `output_dir`
- If google drive is mounted, it will be also saved at `MyDrive/Upscaled Videos (REAL-ESRGAN)`


```python
import os
import glob
import cv2
import subprocess

# --- KONFIGURASI FOLDER & PARAMETER ---
input_folder = "/content/gdrive/MyDrive/Input-gambar/" #@param{type:"string"}
output_dir = "/content/gdrive/MyDrive/hasil-gambar/" #@param{type:"string"}

resolution = "4 x original" # @param ["FHD (1920 x 1080)", "2k (2560 x 1440)", "4k (3840 x 2160)","2 x original", "3 x original", "4 x original"] {type:"string"}
model = "RealESRGAN_x4plus" #@param ["RealESRGAN_x4plus" , "RealESRGAN_x4plus_anime_6B", "realesr-animevideov3"]

# Pastikan folder output ada
os.makedirs(output_dir, exist_ok=True)

# Cari semua file mp4
list_video = glob.glob(os.path.join(input_folder, "*.mp4"))

if len(list_video) == 0:
    print(f"❌ Tidak ada file .mp4 ditemukan di: {input_folder}")
else:
    print(f"✅ Ditemukan {len(list_video)} video. Memulai proses batch...")

# --- MULAI LOOPING ---
for i, video_path in enumerate(list_video):
    video_name_with_ext = os.path.basename(video_path)
    print(f"\n=================================================")
    print(f"🎬 Memproses Video [{i+1}/{len(list_video)}]: {video_name_with_ext}")
    print(f"=================================================")

    # Baca info video
    video_capture = cv2.VideoCapture(video_path)
    video_width = int(video_capture.get(cv2.CAP_PROP_FRAME_WIDTH))
    video_height = int(video_capture.get(cv2.CAP_PROP_FRAME_HEIGHT))
    video_capture.release() # Tutup capture setelah baca info

    final_width = None
    final_height = None
    aspect_ratio = float(video_width/video_height)

    # Tentukan resolusi target
    match resolution:
        case "FHD (1920 x 1080)":
            final_width=1920
            final_height=1080
        case "2k (2560 x 1440)":
            final_width=2560
            final_height=1440
        case "4k (3840 x 2160)":
            final_width=3840
            final_height=2160
        case "2 x original":
            final_width=2*video_width
            final_height=2*video_height
        case "3 x original":
            final_width=3*video_width
            final_height=3*video_height
        case "4 x original":
            final_width=4*video_width
            final_height=4*video_height

    # Logic aspect ratio
    if aspect_ratio == 1.0 and "original" not in resolution:
        final_height = final_width

    if aspect_ratio < 1.0 and "original" not in resolution:
        temp = final_width
        final_width = final_height
        final_height = temp

    # Hitung Scale Factor
    scale_factor = max(final_width/video_width, final_height/video_height)
    isEven = int(video_width * scale_factor) % 2 == 0 and int(video_height * scale_factor) % 2 == 0

    # Scale factor needs to be even
    while isEven == False:
        scale_factor += 0.01
        isEven = int(video_width * scale_factor) % 2 == 0 and int(video_height * scale_factor) % 2 == 0

    print(f"   -> Upscaling from {video_width}x{video_height} to {final_width}x{final_height}")
    print(f"   -> Scale Factor: {scale_factor:.2f}")

    # --- JALANKAN COMMAND REAL-ESRGAN ---
    # Menggunakan f-string di Python command agar lebih aman
    cmd_upscale = f"python Real-ESRGAN/inference_realesrgan_video.py -n {model} -i '{video_path}' -o '{output_dir}' --outscale {scale_factor}"
    os.system(cmd_upscale)

    # --- POST PROCESSING (CROP & RENAME) ---
    video_name = video_name_with_ext.replace(".mp4","")
    upscaled_video_path = f"{output_dir}/{video_name}_out.mp4" # Path output default dari RealESRGAN
    final_video_name = f"{video_name}_upscaled_{final_width}_{final_height}.mp4"
    final_video_path = os.path.join(output_dir, final_video_name)

    # Cek apakah file hasil upscale benar-benar ada sebelum lanjut
    if os.path.exists(upscaled_video_path):
        # crop to fit using ffmpeg
        if "original" not in resolution:
            print("   -> Cropping to fit exact resolution...")
            command = f"ffmpeg -loglevel error -hwaccel cuda -y -i '{upscaled_video_path}' -c:v h264_nvenc -filter:v 'crop={final_width}:{final_height}:(in_w-{final_width})/2:(in_h-{final_height})/2' -c:v libx264 -pix_fmt yuv420p '{final_video_path}'"
            subprocess.run(command, shell=True)
        else:
            # Jika original, cukup rename/copy
            command = f"cp '{upscaled_video_path}' '{final_video_path}'"
            subprocess.run(command, shell=True)

        print(f"✅ Video selesai: {final_video_path}")

        # Hapus file intermediate (_out.mp4) untuk menghemat ruang
        if os.path.exists(upscaled_video_path):
            os.remove(upscaled_video_path)

        # --- COPY KE FOLDER CADANGAN (Opsional sesuai script asli) ---
        # Asumsi variable mount_drive diatur di cell lain, jika tidak ada, kita skip biar ga error
        try:
            if 'mount_drive' in globals() and mount_drive:
                drive_folder = "MyDrive/Upscaled Videos (REAL-ESRGAN)"
                save_directory_drive = f"/content/gdrive/{drive_folder}"
                os.makedirs(save_directory_drive, exist_ok=True)

                command = f"cp '{final_video_path}' '{save_directory_drive}/{final_video_name}'"
                subprocess.run(command, shell=True)
                print(f"   -> Saved copy to drive: /{drive_folder}/{final_video_name}")
        except NameError:
            pass # mount_drive tidak didefinisikan, skip saja

    else:
        print(f"❌ Gagal: File hasil upscale tidak ditemukan untuk {video_name}")

print("\n🎉 SEMUA PROSES BATCH SELESAI!")
```
